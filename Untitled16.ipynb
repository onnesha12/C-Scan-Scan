```python
def scan_disk(requests, head, direction, disk_size=200):
    requests.sort()
    left = [r for r in requests if r < head]
    right = [r for r in requests if r >= head]

    seek_sequence = []
    total_movement = 0
    current_head = head

    if direction == "left":
        # Move left first
        for r in reversed(left):
            total_movement += abs(current_head - r)
            current_head = r
            seek_sequence.append(r)

        # then move to 0
        total_movement += abs(current_head - 0)
        current_head = 0

        for r in right:
            total_movement += abs(current_head - r)
            current_head = r
            seek_sequence.append(r)

    else:

        for r in right:
            total_movement += abs(current_head - r)
            current_head = r
            seek_sequence.append(r)


        total_movement += abs(current_head - (disk_size - 1))
        current_head = disk_size - 1

        for r in reversed(left):
            total_movement += abs(current_head - r)
            current_head = r
            seek_sequence.append(r)

    return total_movement, seek_sequence


def cscan_disk(requests, head, direction, disk_size=200):
    requests.sort()
    left = [r for r in requests if r < head]
    right = [r for r in requests if r >= head]

    seek_sequence = []
    total_movement = 0
    current_head = head

    if direction == "left":
        for r in reversed(left):
            total_movement += abs(current_head - r)
            current_head = r
            seek_sequence.append(r)


        total_movement += abs(current_head - 0)
        total_movement += (disk_size - 1)

        current_head = disk_size - 1
        for r in reversed(right):
            total_movement += abs(current_head - r)
            current_head = r
            seek_sequence.append(r)

    else:
        for r in right:
            total_movement += abs(current_head - r)
            current_head = r
            seek_sequence.append(r)

        total_movement += abs(current_head - (disk_size - 1))
        total_movement += (disk_size - 1)

        current_head = 0
        for r in left:
            total_movement += abs(current_head - r)
            current_head = r
            seek_sequence.append(r)

    return total_movement, seek_sequence


requests = [176, 79, 64, 90, 21, 39]
head = 50
direction = "left"

scan_total, scan_seq = scan_disk(requests, head, direction)
cscan_total, cscan_seq = cscan_disk(requests, head, direction)

print("SCAN Total Head Movement:", scan_total)
print("SCAN Sequence:", scan_seq)
print("C-SCAN Total Head Movement:", cscan_total)
print("C-SCAN Sequence:", cscan_seq)
```
